```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
# Replace with your Oracle credentials and connection details
oracle_username = 'Tanui'
oracle_password = 'Oracle2024'
oracle_host = 'localhost'
oracle_port = '1521'
oracle_service = 'xe'

# Connection string for Oracle
oracle_connection_string = f'oracle+cx_oracle://{oracle_username}:{oracle_password}@{oracle_host}:{oracle_port}/{oracle_service}'
```

```python
# Create a SQLAlchemy engine to connect to Oracle
engine = create_engine(oracle_connection_string)
```

```python
customer_orders = pd.read_csv('../cleanfiles/customer_orders.csv')
customer_orders["Stems Ordered"] = customer_orders["Stems Ordered"].astype('int')
customer_orders["Date"] = pd.to_datetime(customer_orders["Date"])

table_name = 'customer_orders'
customer_orders.to_sql(table_name, con=engine, index=False, if_exists='replace')

customer_orders
```
